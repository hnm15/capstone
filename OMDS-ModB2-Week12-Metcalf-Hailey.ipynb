# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

Plot 1: The advantages of this graph is that each week is shown on its own graph; there isn't overlap so it is easier to see the specific points and trends for each day within the two graphs. In addition, because the graphs are side by side, it is easier to see the whole picture or the trend over the two weeks. We can see week 2 started off lowers than week 1 with relative easy for the viewer. A disadvantage of this graph is it can be a bit difficult to read due to the two y axises. In addition, it is harder to directly compare the days to each other - the viewer has to jump back and forth between the graphs for a direct comparison.

Plot 2: The advantages of this graph are that both weeks are one plot. The ability to compare the days to each other is easy and can provide a quick insight into the trends when comparing the two weeks. There is also only 1 x axis and 1 y axis for this graph and the two weeks are different colors so there is not a ton of confusion (or extra knowledge required) for the viewer. The disadvantage of this graph is the points are not as exact and there is some overlap; the information is not as precise as plot 1.
